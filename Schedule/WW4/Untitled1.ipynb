```python
def MAJ(a,b,c):
    return NAND(NAND(NAND(NAND(a,b),NAND(a,c)),NAND(NAND(a,b),NAND(a,c))),NAND(b,c))

def MAJ2(a,b,c):
    return OR(AND(a,b), OR(AND(b,c), AND(a,c)))
```

```python
circuit(MAJ).draw()
circuit(MAJ2).draw()
```
